```python
tois.columns
```

```python
toi = 685.01
idx = tois['TOI']==toi
ra,dec = tois.loc[idx, ['RA','Dec']].values[0]
```

```python
from astropy.coordinates import SkyCoord
from astropy import units as u

ra_dec = SkyCoord(' '.join((ra,dec)), unit=[u.hourangle, u.degree])
```

```python
from glob import glob

files = glob('../scripts/k2-261/ADP.2018-01-26T01:03:21.955/data/reduced/2018-01-24/*.fits')
files.sort()
files
```

```python
%matplotlib inline
import matplotlib.pyplot as pl
from astropy.io import fits

file = files[0]
with fits.open(file) as f:
    print(f.info())
    header = f[0].header
    data = f[0].data
data.shape
```

```python
header['RA'], ra_dec.ra.degree
```

```python
assert (header['RA']*u.degree-ra_dec.ra)<10*u.arcsec
```

```python
assert (header['Dec']*u.degree-ra_dec.dec)<10*u.arcsec
```

```python
pl.matshow(data)
```
